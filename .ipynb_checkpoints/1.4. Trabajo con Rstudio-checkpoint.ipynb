# 1. Instalación de Rstudio

Como ya hemos dicho en las lecciones anteriores, a veces utilizarás trabajarás con R usando estas lecciones y otras veces usarás un programa llamado [Rstudio](https://rstudio.com/).

Rstudio es un programa muy versatil y es lo que todos los que trabajamos en R utilizamos en el día a día. Es un programa comercial pero dispone de una versión llamada Desktop que es (y siempre será) gratuita para los usuarios.

Dispone de Rstudio completamente instalado en los ordenadores de la facultad, pero suele ser cómodo tenerlo también instalado en el portátil o en el ordenador de casa, para ello sigue estos pasos:

## Instrucciones para Windows:
*** 

1. Descárgate R en tu ordenador desde este [enlace](https://cran.rediris.es/bin/windows/base/R-3.6.2-win.exe). Ese es el paquete oficial de R que debes instalar en tu ordenador, si requieres más información ve a la página oficial de R: [https://www.r-project.org](https://www.r-project.org)

3. Descárgate Rstudio desde este [enlace](https://download1.rstudio.org/desktop/windows/RStudio-1.2.5033.exe)

4. Instala Rstudio, igual que antes no cambies la configuración si no sabes qué estás haciendo.

5. Abre Rstudio simplemente clicando dos veces en el icono que te habrá aparecido.

## 2. Un primer vistazo a Rstudio

Al arrancar Rstudio (clicando dos veces en el icono) os abrirá el programa y veréis que está dividido en 4 ventanas llamadas: source, console, environment y files.

* source: Es donde se escribe el código.
* console: es la consola de R, donde se producen los resultados de los análisis.
* environment: información de datos y otras variables.
* files: ficheros, ayuda, plots y otras cosas.

Se puede cambiar la organización de las ventanas en __Tools --> Global Options --> Pane Layout__, por ejemplo para mí tiene más sentido esta organización:

<img src="imagenes/1.4.panelayout.png" style="width: 600px;">

_Source_ arriba a la izquiera, _Environment_ abajo a la izquierdam _Console_ arriba a la derecha y _Files_ abajo a la derecha, ahora te explicaré el porqué.

En Rstudio el código se escribe en la ventana de _Source_. El primer código que se introduce en cualquier lenguaje de programación es siempre _"Hello world"_ ([por si quieres saber por qué](https://hipertextual.com/2016/01/hello-world-origen)) y es lo que vamos a hacer, escribir nuestro primer "Hello world".

Vamos a la ventona de _source_ y escribimos:

```R
print("Hello world")
```
Y ahora para ejecutarlo lo seleccionamos y pulsamos __Control + Enter__

Verás que en la ventana de la consola te habrá aparecido lo siguiente:

```
> print("Hello world")
[1] "Hello world"
```

Y así se trabaja en Rstudio, escribes el código en source, lo ejecutas con __Control + Enter__ y ves los resultados en la consola, sencillo. Por eso a mí me gusta la configuración de ventanas que te he dicho antes, escribo el código en la izquierda y veo los resultados en la derecha, de esa manera optimizo el espacio que utilizo de la pantalla y me resulta más cómodo trabajar.

Además, cuando tengas una porción de código que quieras salvar puedes hacerlo en __File --> Save As__ y nunca más tendrás que volver a escribir ese código ni perder tiempo repitiendo esos análisis, simplemente abres ese fichero y lo ejecutas.

__Consejo:__ Al principio todo esto puede que te resulte confuso, no te preocupes, iremos practicando en clase y al final le cogerás el truco.


# 3. Otras formas de cargar datos

<p style="font-size:130%;font-family:verdana;color:red;">Lee esto con detenimiento porque  te puede salvar la vida en alguna práctica</p>


En la lección anterior hemos visto que para vargar los datos teníamos que utilizar la función _read.csv()_. Cuando estamos trabajando con __Rstudio__ podemos también leer los datos utilizando un menú que hay en la ventana de objetos, pincha en __import Dataset__ y selecciona __From Text (base)__ te abrirá el típico menú para que selecciones el fichero que quieres abrir. 

<img src="imagenes/1.3.selectdataset.png" style="width: 800px;">


Una vez seleccionado se abrirá un menú como el siguiente, en _Name_ pon el nombre que quieres para tus datos (pe, mydata, misdatos, lo que quieras) y en _Heading_ marca _"yes"_. Lo más seguro es que R detecte correctamente la configuración, pero por si acaso compruébalo (Separator = semicolon y Decimal = Comma), dale a aceptar y tendrás los datos cargados con el nombre que hayas puesto en _Name_.

> Las imágenes y explicaciones que he dado son para el lenguaje inglés (que es la versión que tengo instalada) si tienes una versión en español, no te preocupes, es exactamente igual.


__Importante:__ La forma de leer ficheros que acabo de explicar __es una mala opción__. En R queremos tener los procesos de análisis automatizados (esa es una gran ventaja respecto a SPSS) y cargando los datos de esta manera tenemos que perder tiempo abriendo los ficheros. No pasa nada si la utilizamos al principio, hasta que nos sintamos cómodos con R, pero poco a poco la tendremos que sustituir por el uso de la sintaxis.


# 4. El directorio de trabajo en R

Todos los programas, cuando se arrancan establecen un directorio de trabajo, es importante que la entiendas para no volverte loco cuando trabajes con __Rstudio__.

Cuando abres R el programa, éste establece un directorio de trabajo. Todas las acciones que hagas (leer datos, escribir datos, guardar imágenes) siempre se hacen tomando como referencia ese directorio de trabajo.

Por ejemplo, suponte que tienes la siguiente estructura de ficheros en tu ordenador:

```
Practica01/
  codigo.R
  datos/
    basededatos.csv
```

Es decir, tienes una carpeta llamada __Practica01__, dentro de ella tienes un fichero llamado _codigo.R_ (que contiene tu código) y una carpeta llamada __datos__ que contiene el fichero _basededatos.csv_ (que son tus datos).

Cuando vayas a empezar a trabajar, al clicar dos veces sobre _codigo.R_ vas a abrir Rstudio, y vas a hacer que ese sea tu directorio de trabajo.

Ahora para leer los ficheros tendrás que usar la sintaxis:

```R
    mydata <- read.csv(file="datos/basededatos.csv", header=TRUE, sep=";", dec =",")
``` 

Y si tu estructura de ficheros fuese como esta (no tienes la carpeta datos):

```
Practica01/
  codigo.R
  basededatos.csv
```

Escribirías:

```R
    mydata <- read.csv(file="basededatos.csv", header=TRUE, sep=";", dec =",")
``` 

Y un último ejemplo, si tu estructura de datos es:

```
Practica01/
  codigo.R
  datos/
    test1/
      basededatos.csv
```

Tendrías que poner:


```R
    mydata <- read.csv(file="datos/test1/basededatos.csv", header=TRUE, sep=";", dec =",")
``` 

***

__Resumiendo:__ Cuando uses ficheros siempre tienes que tener en cuenta cuál es tu directorio de trabajo, e indicar el nombre del fichero junto con el _path_ (el camino) hasta tu fichero.

En R cuando quieras saber cuál es tu directorio de trabajo puedes escribir:

```R
    getwd()
``` 

***

Y si quieres cambiar el directorio de trabajo en Rstudio puedes ir en la ventana de Files hasta la carpeta que deseas, y una vez en ella hacer en el menú de la ventana: __More --> Set As Working Directory__

<img src="imagenes/1.4.setworkingdirectory.png" style="width: 600px;">

***

Un truco para evitarte problemas con los directorios de trabajo en las prácticas es el siguiente:

* 1. Crea una carpeta (pe: Practica01)
* 2. Guarda tus datos directamente en esa carpeta
* 3. Crea un fichero llamado codigo.R
* 4. Clica dos veces en codigo.R y así siempre sabrás donde están tus datos y podrás acceder a ellos son un simple:

```R
    mydata <- read.csv(file="basededatos.csv", header=TRUE, sep=";", dec =",")
``` 
***
__No te preocupes si esto te ha resultado algo confuso (no lo es, en realidad así es cómo funcionan todos los ordenadores) en clase lo explicaré con nuevos ejemplos.__

***

<div align="right">
  Enero de 2020 |
  <a href="autores.html"> AUTORES <a>
</div>
