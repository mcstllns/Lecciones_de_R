# 2. R como calculadora numérica

# 3. Conceptos de variable y objeto

Una vez que tenemos claro que R puede hacer cualquier tipo de cálculo, hemos de entender cómo nuestro Software va a tratar dichos cálculos. Aquí aparecen dos conceptos fundamentales: __variable y objeto__

__Un objeto__ es cualquier elemento de R, sea cual sea su naturaleza (una gráfica, una base de datos, funciones, o un simple número); y __una variable__ es un tipo de objeto que puede tomar diferentes valores. 
Ahora bien, veamos qué podemos hacer con esas variables.

Algo que podemos hacer es crear una variable llamada “a” y asignarle el número  5. La sintaxis de R que hace esto es la siguiente

```R 
     a <- 5 
```

Ahora la variable a está guardada en la memoria de R con un valor de __5__ . Fíjate que hemos utilizado el símbolo "<-" en vez del igual "=". El primero se interpreta como una flecha y tiene el significado de __asignar__ un valor a algo. En R se prefiere el símbolo de la flecha al del igual.

Ahora, si queremos comprobar que efectivamente tiene ese valor podemos pedirle a R que nos muestre su valor con la función _print()_.

```R 
     print(a) 
```

O, lo que es mucho más cómodo, escribiendo simplemente el nombre de la variable y ejecutando el código


```R 
     a
```

# 3. Como acceder a elementos de un vector

Como hemos dicho antes, el concepto clave en R es el de objeto y ahora vamos a ver cómo funcionan y cómo podemos crearlos y acceder a aquellos valores que lo forman. 

Tened en cuenta que en esta explicación vamos a manejar variables que nosotros mismos crearemos para facilitar el entendimiento del código, pero que este procedimiento se aplicaráa exactamente igual cuando trabajemos con datos que leeamos de un archivo.

Lo primero será crear una variable, por ejemplo, Imaginemos que hemos pasado un test de Inteligencia (_CI_) a 10 aspirantes a un puesto de trabajo y los valores obtenidos han sido los siguientes:

```R 
     ci <- c(120,123,125,90,160,134,110,100,170,145)
```

Con la sintaxis anterior hemos creado un vector formado por 10 elementos, donde el primero es 120, el segundo 123 y así hasta el último (el décimo) cuyo valor es 145. Si quiero comprobar que R lo ha guardado correctamente simplemente escribo su nombre y veo los elementos que componen el vector _ci_.

```R 
     ci
```

Supongamos que estos datos han sido ordenados por orden de llegada a la prueba, y queremos saber el resultado en CI del participante número 3, para eso utilizaríamos:

```R 
     ci[3]
```

Poniendo el número del elemento entre corchetes podemos acceder al valor del elemento que está en la tercera posición. Otra opción sería acceder a varios datos al mismo tiempo, supongamos que nos interesa saber los valores de los primeros tres participantes:

```R 
     ci[1:3]
```

Esta es una forma muy típica en R, la expresión 1:3 le dice a R "_todos los números entre el 1 y el 3_". Por otro lado, si lo que quiero es acceder a elementos no consecutivos se puede utilizar la siguiente forma. Si necesitamos acceder a los datos del 4 y el 10, en ese caso habríamos de usar:

```R 
     ci[c(4,10)]
```

Ahora que hemos entendido la forma básica para acceder a elementos de un vector, paseemos a algo un poco más avanzado y que es más útil. La realidad es que la mayoría de las veces que deseemos acceder a un elemento de un objeto de R lo haremos con una condición, es decir, buscando los datos que cumplan alguna condición que nos sea interesante.

Imaginemos que deseamos reducir la lista de participantes al puesto, y queremos acceder a los datos de aquellos participantes que tengan un CI superior a 120, para ello:

```R 
    ci[ci > 120]
```

Fíjate que nos devuelve los valores: 123, 125, 160, 134, 170 y 145. Todos ellos superiores a 120. Si en vez de querer los valores quisiéramos saber qué elementos cumplen esa condición haríamos:

```R 
    which(ci>120)
```

Esta función nos daría la posición que ocupan en el objeto los datos que encajan con nuestra condición (en este caso los valores en las posiciones 2 3 5 6 9 10); ahora podríamos crear una nueva variable (_listos_) que contenga solamente los candidatos que han superado este primer filtro:

```R 
    listos <- which(ci>120)
```

Y ahora para mostrar los contenidos de esta variable tenemos las dos opciones que ya conocemos, o simplemente escribimos el nombre de la variable o usamos la función _print()_

```R 
    listos
    print(listos)
```

> MUY IMPORTANTE: Fíjate que listos contiene las posiciones de los que tienen más de 120 de CI, pero no sus valores, para tener los valores tendríamos que hacer:

```R 
    listosCI <- ci[ci>120]
```

En R hay que tener en cuenta que los comparadores son los siguientes:

* $>$ mayor que 
* $<$ menor que
* $>=$ mayor o igual que
* $<=$ menor o igual que
* $==$ igual que

# 4. Los data.frame

Si recuerdas algo de SPSS, los datos se organizaban en forma de filas y columnas, como si fuese una __hoja de datos__ de excel. Esa estructura en R se llama _data frame_ y es exactamente igual a lo visto con el SPSS: __en las columnas están las variables y en las filas están los sujetos__.

Estos data frames normalmente son datos que se leen de un archivo pero como R es un lenguaje de programación podemos crear uno muy sencillo para verlo de ejemplo.

¿Recuerdas que habíamos creado las variables CI y sexo? puedo construir un data.frame con ellas que se llame _misdatos_

```R 
    misdatos <- data.frame(ci, sexo)
```
Y al igual que antes para visualizarlo escribimos su nombre

```R 
    misdatos
```
